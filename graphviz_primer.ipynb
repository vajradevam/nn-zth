````python
import graphviz

dot = graphviz.Digraph('round-table', comment="The Rround Table")

dot.node('A', 'apple')
dot.node('B', 'Pineapple')
dot.edge('apple', 'pineapple')

dot
````

````python
dot.graph_attr['rankdir'] = 'LR'  
dot.edge_attr.update(arrowhead='vee', arrowsize='1')

dot.edge('pineapple', 'lalu grapes')
````

````python
from graphviz import Digraph
````

````python
dot = Digraph(format='svg', graph_attr={'rankdir': 'LR'}) 
dot.edge('a', 'b')
dot
````

````python
dot = Digraph(format='svg', graph_attr={'rankdir': 'LR'}) 
edges = (('a', 'b'), ('b', 'c'), ('c', 'd'))
dot.edges(edges)
dot
````

````python
dot = Digraph(format='svg', graph_attr={'rankdir': 'LR'}) 

for n in range(5):
    uid = str(id(n))
    dot.node(name = uid, label = "{%s}" % (chr(n + 65)), shape = 'record')

dot
````
